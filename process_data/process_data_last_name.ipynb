```python
import pandas as pd
import numpy as np
import math
import os
```

```python
path = os.path.abspath(os.getcwd())
df = pd.read_csv(path + "/Data2/Last Names0.csv")
df = df.dropna(axis = 0, how = 'all')
df = df.where(df.notnull(), "NULL")
print(df.columns)
df.head()
```

```python
# create dictionary like : {'马': {'id': 1, 'pinyin':'Ma'}, '李': ...}
name_pinyin_id = {}
count = 1
name_heading = list(df.columns)[3:8]
pinyin_heading = list(df.columns)[8:13]

# iterate df to create "name_pinyin_id" dictionary
for i in range(len(df)):
    for j in range(len(name_heading)):
        name = df[name_heading[j]][i]
        pinyin = df[pinyin_heading[j]][i]
        if name not in name_pinyin_id and name != "NULL":
            name_pinyin_id[name] = {}
            name_pinyin_id[name]['id'] = count
            name_pinyin_id[name]['pinyin'] = pinyin
            count = count + 1
            
# replace name with id in df
for name in name_pinyin_id:
    df = df.replace(name, name_pinyin_id[name]['id'])

df.head(1)
```

```python
name_pinyin_id
```

```python
# create dataframe for lastName table
df_last_name = pd.DataFrame(columns = ['gazetteerId', 'totalNumber', 'firstNameId', 'secondNameId', 'thirdNameId', 'fourthNameId','fifthNameId'])

# get corresponding original df heading
df_columns = [df.columns[0], df.columns[2]]
df_columns.extend(name_heading)
print(df_columns)

# assign df corresponding columns to df_last_name
for i in range(len(df_columns)):
    df_last_name[df_last_name.columns[i]] = df[df_columns[i]]

df_last_name.head()
```

```python
# create category_df
category_df = pd.DataFrame(columns = ['id', 'name', 'pinyin'])
dic_for_category_df = {'id':[], 'name':[], 'pinyin':[]}

for name in name_pinyin_id:
    dic_for_category_df['name'].append(name)
    dic_for_category_df['id'].append(name_pinyin_id[name]['id'])
    dic_for_category_df['pinyin'].append(name_pinyin_id[name]['pinyin'])
        
for attribute in category_df.columns:
    category_df[attribute] = dic_for_category_df[attribute]

category_df.head()
```

```python
df_last_name.to_csv('last_name_df.csv', index = False, na_rep = "NULL")
```
